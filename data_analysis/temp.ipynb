```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load the dataset
dataset = pd.read_csv('/home/mohamed/repos/nlp_proj/train-cleaned-corrected.csv')

# First, split the data into a temporary train set and a test set (70% train, 30% test)
train_data, test_data = train_test_split(dataset, test_size=0.3, random_state=42)

# Then, split the test_data into validation and test sets (50% validation, 50% test)
val_data, test_data = train_test_split(test_data, test_size=0.5, random_state=42)

# Now you have train_data (70%), val_data (15%), and test_data (15%)

# Save the splits to CSV files
train_data.to_csv('/home/mohamed/repos/nlp_proj/train_split.csv', index=False)
val_data.to_csv('/home/mohamed/repos/nlp_proj/val_split.csv', index=False)
test_data.to_csv('/home/mohamed/repos/nlp_proj/test_split.csv', index=False)

print("Data has been split and saved successfully!")
```
